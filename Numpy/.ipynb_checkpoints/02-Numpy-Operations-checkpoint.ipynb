```python
import numpy as np
```

### Arithmetic

```python
arr = np.arange(10)
```

```python
arr
```

```python
arr + arr
```

```python
arr * arr
```

```python
arr - arr
```

```python
arr / arr
```

```python
1 / arr
```

```python
arr ** 3
```

### Array Functions

```python
# Taking Square Roots
np.sqrt(arr)
```

```python
# Calculating exponential (e^)
np.exp(arr)
```

```python
# Trigonometric Functions like sine
np.sin(arr)
```

```python
# Taking the Natural Logarithm
np.log(arr)
```

### Summary Statistics on Arrays

```python
arr = np.arange(0,10)
arr
```

```python
arr.sum()
```

```python
arr.mean()
```

```python
arr.max()
```

```python
arr.min()
```

```python
arr.var() # variance
```

```python
arr.std() # standard deviation
```

### Axis Logic

```python
arr_2d = np.arange(25).reshape(5, 5)
arr_2d
```

```python
arr_2d.sum(axis=0)
```

```python
arr_2d.sum(axis=1)
```
